### Using CSV file

```python
import pandas as pd
data = pd.read_csv('houseprice.csv')
data[:10]
```

```python
price=[]
#age : age of the house
#nb : Number of bedrooms in the house
#sm : Total area of the house (square meters)

for (age, nb, sm) in zip(data['Age'], data['Bedroom'], data['Area']):
    
    if sm>160:
       
        if nb>3:
           
            if age<10:
                price.append(510)
            else:
                price.append(382)
        else:
            
            if age<40:
                price.append(335)
            else:
                price.append(312)
    else:
       
        if age<25:
           
            if nb>2:
                 price.append(341)
            else:
                 price.append(258)
        else:
            if sm>100:
                 price.append(285)
            else:
                 price.append(217)
    
```

```python
data['house Price']=price
```
